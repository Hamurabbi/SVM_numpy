```python
import numpy as np
from src.SVM import SVM
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
import seaborn as sns
sns.set()
```

```python
X, y = make_blobs(n_samples=20, centers=2,cluster_std=0.60)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn');
```

```python
model = SVM(learning_rate = 1, epochs = 100)
```

```python
w, out = model.fit(X,y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn');

# Print the hyperplane calculated by
x2=[w[0],w[1],-w[1],w[0]]
x3=[w[0],w[1],w[1],-w[0]]

x2x3 =np.array([x2,x3])
X1,Y1,U,V = zip(*x2x3)
ax = plt.gca()
ax.quiver(X1,Y1,U,V,scale=1, color='blue')
```
